```python
import pandas as pd
import numpy as np
```

```python
train_data=pd.read_csv("Z:\\AV\\club mahindra\\train_5CLrC8b\\train.csv")
test_data=pd.read_csv("Z:\\AV\\club mahindra\\test_Jwt0MQH\\test.csv")
```

```python
train_data['booking_date']=pd.to_datetime(train_data.booking_date,infer_datetime_format=True)
train_data['checkin_date']=pd.to_datetime(train_data.checkin_date,infer_datetime_format=True)
train_data['checkout_date']=pd.to_datetime(train_data.checkout_date,infer_datetime_format=True)
test_data['booking_date']=pd.to_datetime(test_data.booking_date,infer_datetime_format=True)
test_data['checkin_date']=pd.to_datetime(test_data.checkin_date,infer_datetime_format=True)
test_data['checkout_date']=pd.to_datetime(test_data.checkout_date,infer_datetime_format=True)
```

```python
date_difference_train=pd.DataFrame(abs(train_data.booking_date-train_data.checkin_date))
date_difference_test=pd.DataFrame(abs(test_data.booking_date-test_data.checkin_date))
```

```python
date_difference_train=(date_difference_train / np.timedelta64(1, 'D')).astype(int)
date_difference_test=(date_difference_test / np.timedelta64(1, 'D')).astype(int)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler(feature_range=(0,1))
train_data['difference']=scaler.fit_transform(date_difference_train)
test_data['difference']=scaler.fit_transform(date_difference_test)

```

```python
train_data[train_data.columns[0:]].corr()['amount_spent_per_room_night_scaled'][:]
```

```python
train_data=train_data.fillna(method='ffill',axis=0)
test_data=test_data.fillna(method='ffill',axis=0)

```

```python
train_data.dtypes
```

```python
features=['channel_code','main_product_code','numberofadults','persontravellingid','numberofchildren','persontravellingid','resort_type_code','resort_region_code','room_type_booked_code','roomnights','season_holidayed_code','total_pax',
         'state_code_resort','difference','cluster_code','member_age_buckets','reservationstatusid_code','resort_id']
X=train_data[features]
y=train_data['amount_spent_per_room_night_scaled']
```

```python
X.columns
```

```python
from sklearn.model_selection import train_test_split
X_train,X_valid,y_train,y_valid=train_test_split(X,y,test_size=0.4,random_state=1)
```

```python
cat_features=[14,15,16,17]
```

```python
from catboost import CatBoostRegressor
my_model = CatBoostRegressor()
my_model.fit(
    X_train, y_train,cat_features
  #  eval_set=(X_valid, y_valid),
#     logging_level='Verbose',  # you can uncomment this for text output
)
```

```python
from sklearn.metrics import mean_squared_error
pred=my_model.predict(X_valid)
mean_squared_error(y_valid,pred)
```

```python
test=test_data[features]
prediction=my_model.predict(test)
```

```python
register_id=test_data['reservation_id']
```

```python
submissin = pd.DataFrame({'reservation_id': register_id,
                           'amount_spent_per_room_night_scaled':prediction
                           })
```
